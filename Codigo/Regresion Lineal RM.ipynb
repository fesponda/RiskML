```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from random import random, uniform
```

```python
df = pd.read_csv("/Users/fesponda/Itam/Itam/Clases Itam/RiskMath/RiskDistrib/Datos/regLin.csv")
```

```python
X_train, X_test, y_train, y_test =train_test_split(df[['X']],df['y'],test_size=0.25)
```

Convertir a arreglo despues de hacer toda la limpieza de datos, integracion etc

```python
X_train=np.asarray(X_train)
X_test=np.asarray(X_test)
y_train=np.asarray(y_train)
y_test=np.asarray(y_test)
```

## Visualizar

```python
plt.scatter(X_train,y_train)
```

```python
modelo=LinearRegression()
```

```python
modelo.fit(X_train,y_train)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

## Visualizar la recta

```python
x=np.asarray([[i] for i in range(100)])
y=[modelo.intercept_ +modelo.coef_[0]*i for i in x]
#y=modelo.predict(x)
```

```python
plt.scatter(X_train,y_train)
plt.plot(x,y,c='r')
```

Visualizar con datos de prueba

```python
x=np.asarray([[i] for i in range(100)])
y=modelo.predict(x)
plt.scatter(X_test,y_test)
plt.plot(x,y,c='r')
```

```python
def error(w0,w1,X,y_train):
    y=[w0 +w1*i[0] for i in X]
    return np.mean((y_train-y)**2)
```

 Visualizar como cambia el error conforme cambian las w's. Por ejemplo w1

```python
error(modelo.intercept_,modelo.coef_[0],X_train,y_train)
```

```python
w=[random() for i in range(100)]
```

```python
w=sorted(w)
```

```python
errores=[error(modelo.intercept_,k,X_train,y_train) for k in w]
```

```python
plt.plot(w,errores)
plt.scatter(modelo.coef_[0],error(modelo.intercept_,modelo.coef_[0],X_train,y_train),c='r')
```
